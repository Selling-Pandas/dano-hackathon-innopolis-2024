```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
import scipy
import statsmodels.api as sm

%matplotlib inline
%config InlineBackend.figure_format = 'retina'

plt.rcParams['figure.figsize'] = 14, 6
plt.rcParams['font.size'] = 18
plt.rcParams['savefig.format'] = 'pdf'
plt.rcParams['font.family'] = 'hse sans'
plt.rcParams['font.weight'] = 'semibold'
sns.set_style('darkgrid')
```

```python
df = pd.read_csv('df_extended.csv').drop(columns='Unnamed: 0')
```

```python
def formatting_first(x):
    if x == 'ПОЛНОСТЬЮ УДОВЛЕТВОРЕНЫ':
        return 5
    elif x == 'СКОРЕЕ УДОВЛЕТВОРЕНЫ':
        return 4 
    elif x == 'И ДА, И НЕТ':
        return 3
    elif x == 'СКОРЕЕ НЕ УДОВЛЕТВОРЕНЫ':
        return 2
    elif x == 'СОВСЕМ НЕ УДОВЛЕТВОРЕНЫ':
        return 1
    return np.nan


def formatting_second(x):
    if x == 'НИЗШАЯ СТУПЕНЬ':
        return 1
    elif x == 'ВЫСШАЯ СТУПЕНЬ':
        return 9
    elif x == 'ЗАТРУДНЯЮСЬ ОТВЕТИТЬ' or x == 'ОТКАЗ ОТ ОТВЕТА' or x == 'НЕТ ОТВЕТА':
        return np.nan
    return x
```

```python
df['vj1.1.2'] = df['vj1.1.2'].apply(formatting_first)
df['vj1.1.3'] = df['vj1.1.3'].apply(formatting_first)
df['vj1.1.4'] = df['vj1.1.4'].apply(formatting_first)
df['vj62'] = df['vj62'].apply(formatting_second).astype(float)
df['vj63'] = df['vj63'].apply(formatting_second).astype(float)
df['vj64'] = df['vj64'].apply(formatting_second).astype(float)
df.dropna(inplace=True)
```

Вывод из следующего линрега таков, что эти переменные - хуета.

```python
predictor = df[['age', 'is_male', 'is_gov', 'internet', 'num_children', 'is_urban', 'education_level', 'vj1.1.2', 'vj1.1.3', 'vj1.1.4', 'vj62', 'vj63', 'vj64']]
linreg = sm.OLS(df['log_wage'], sm.add_constant(predictor))
linreg.fit().summary()
```
